## K-Means Algorithm

### AI/ML Algorithm Profile

| Feature | Description |
| :--- | :--- |
| **AI/ML Category** | Unsupervised Machine Learning |
| **Type** | Clustering Algorithm |
| **Use in Compression** | Color quantization / Image compression |

### What is K-Means?

**K-Means** is an algorithm that **groups similar things together**. It's like organizing a messy room by putting similar items in the same box.



#### 🧱 Real-Life Analogy:
Imagine you have a large box of mixed LEGO bricks:
1. **Step 1:** Randomly pick $K$ bricks to act as your "centers" (e.g., 3 bricks: red, blue, and yellow).
2. **Step 2:** Put every other brick into the group of the center it is most similar to.
3. **Step 3:** Find the "average" brick in each newly formed group (this becomes the new center).
4. **Step 4:** Repeat steps 2 and 3 until no brick needs to move to another group.

#### 🖼️ In Image Compression:
Instead of storing millions of unique colors for a single photograph, K-Means analyzes the image and finds the **$K$ most representative colors** (the centers). It then maps every single pixel in the image to its closest representative color, drastically reducing the file size.

```python
'''

Original: [RGB, RGB, RGB, RGB, RGB, RGB, ...] (millions of colors)
     ↓
K-Means finds 16 representative colors
     ↓
Each pixel stored as: [0, 5, 2, 0, 1, 3, ...] (index to color palette)
     ↓
Compressed: 16 colors + index map

'''
```

#### Objective Function (to minimize):
$$J(V) = \sum_{i=1}^{c} \sum_{j=1}^{c_i} ||x_{ij} - v_i||^2$$

**Where:**
* $x_{ij}$ = data points
* $v_i$ = cluster centers
* $c$ = number of clusters
* $c_i$ = number of points in cluster $i$
* $||x - v||$ = Euclidean distance

---

#### Center Update Formula:
$$v_i = \frac{1}{c_i} \sum_{j=1}^{c_i} x_{ij}$$

*The new center is simply the **mean** of all points in the cluster.*

### Algorithm Steps

```python
'''START
   ↓
1. Randomly select c cluster centers
   ↓
2. Calculate distance between each data point and each center
   ↓
3. Group data based on minimum distance
   ↓
4. Recalculate center positions using equation v_i = (1/c_i)∑x_ij
   ↓
5. Recalculate distances
   ↓
6. Did any point move? ──YES──→ Go to Step 3
   ↓ NO
   ↓
END'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

# ============================================================================
# K-MEANS ALGORITHM IMPLEMENTATION
# ============================================================================

class KMeans:
    """
    K-Means Clustering Algorithm
    
    Parameters:
    - k: number of clusters
    - max_iters: maximum iterations
    - tol: tolerance for convergence
    """
    
    def __init__(self, k=3, max_iters=100, tol=1e-4):
        self.k = k
        self.max_iters = max_iters
        self.tol = tol
        self.centers = None
        self.labels = None
        self.history = []  # Store center history for visualization
        
    def initialize_centers(self, X):
        """Step 1: Randomly select c cluster centers"""
        n_samples = X.shape[0]
        indices = np.random.choice(n_samples, self.k, replace=False)
        return X[indices].copy()
    
    def calculate_distances(self, X, centers):
        """Step 2 & 5: Calculate distance between all points and each center"""
        distances = np.zeros((X.shape[0], self.k))
        
        for i in range(self.k):
            # Euclidean distance: ||x - center||²
            distances[:, i] = np.sum((X - centers[i]) ** 2, axis=1)
        
        return distances
    
    def assign_clusters(self, distances):
        """Step 3: Group based on minimum distance"""
        return np.argmin(distances, axis=1)
    
    def update_centers(self, X, labels):
        """Step 4: Recalculate center positions using equation (8)"""
        new_centers = np.zeros((self.k, X.shape[1]))
        
        for i in range(self.k):
            points_in_cluster = X[labels == i]
            if len(points_in_cluster) > 0:
                # v_i = (1/c_i) * Σ x_ij
                new_centers[i] = np.mean(points_in_cluster, axis=0)
            else:
                # Handle empty cluster - reinitialize randomly
                new_centers[i] = X[np.random.choice(X.shape[0])]
        
        return new_centers
    
    def has_converged(self, old_centers, new_centers):
        """Check if centers stopped moving"""
        return np.sum((new_centers - old_centers) ** 2) < self.tol
    
    def fit(self, X):
        """
        Fit K-Means to data X
        
        Args:
            X: data array of shape (n_samples, n_features)
        """
        print(f"\n🔷 Fitting K-Means with k={self.k}")
        
        # Step 1: Randomly select c cluster centers
        self.centers = self.initialize_centers(X)
        self.history = [self.centers.copy()]
        
        for iteration in range(self.max_iters):
            # Step 2: Calculate distances
            distances = self.calculate_distances(X, self.centers)
            
            # Step 3: Assign based on minimum distance
            self.labels = self.assign_clusters(distances)
            
            # Step 4: Recalculate center positions
            new_centers = self.update_centers(X, self.labels)
            self.history.append(new_centers.copy())
            
            # Step 6: Check convergence (no object moved)
            if self.has_converged(self.centers, new_centers):
                print(f"   Converged after {iteration + 1} iterations")
                break
            
            self.centers = new_centers
            
            if iteration % 20 == 0:
                print(f"   Iteration {iteration + 1}/{self.max_iters}")
        
        # Calculate objective function J(V)
        distances = self.calculate_distances(X, self.centers)
        min_distances = np.min(distances, axis=1)
        self.J = np.sum(min_distances)
        
        print(f"   Final objective J(V) = {self.J:.4f}")
        
        return self
    
    def predict(self, X):
        """Assign new points to nearest cluster"""
        distances = self.calculate_distances(X, self.centers)
        return self.assign_clusters(distances)


# ============================================================================
# K-MEANS FOR IMAGE COMPRESSION
# ============================================================================

class KMeansImageCompression:
    """
    K-Means for Image Compression (Color Quantization)
    """
    
    def __init__(self, k=16):
        self.k = k
        self.kmeans = KMeans(k=k)
        self.palette = None
        
    def compress(self, image_array):
        """
        Compress image using K-Means color quantization
        
        Args:
            image_array: shape (height, width, channels)
            
        Returns:
            compressed: image with k colors
            ratio: compression ratio
            labels: cluster indices
        """
        original_shape = image_array.shape
        h, w, c = original_shape
        
        # Reshape to (pixels, RGB)
        pixels = image_array.reshape(-1, c).astype(np.float32)
        
        print(f"\n📊 Original image: {h}×{w}, {c} channels")
        print(f"   Total pixels: {h*w}")
        print(f"   Unique colors: {len(np.unique(pixels, axis=0))}")
        
        # Fit K-Means to pixel colors
        self.kmeans.fit(pixels)
        self.palette = self.kmeans.centers.astype(np.uint8)
        
        # Assign each pixel to nearest center
        labels = self.kmeans.predict(pixels)
        
        # Reconstruct using palette colors
        compressed = self.palette[labels].reshape(original_shape).astype(np.uint8)
        
        # Calculate compression ratio
        # Original: h*w*c*8 bits
        # Compressed: palette (k*c*8) + indices (h*w*log2(k))
        original_bits = h * w * c * 8
        palette_bits = self.k * c * 8
        index_bits = h * w * np.ceil(np.log2(self.k))
        compressed_bits = palette_bits + index_bits
        
        ratio = original_bits / compressed_bits
        
        print(f"\n✅ Compression complete:")
        print(f"   Colors used: {self.k}")
        print(f"   Palette bits: {palette_bits:.0f}")
        print(f"   Index bits: {index_bits:.0f}")
        print(f"   Compression ratio: {ratio:.3f}:1")
        
        return compressed, ratio, labels.reshape(h, w)
    
    def get_palette_image(self):
        """Return color palette as an image"""
        palette_img = self.palette.reshape(1, self.k, 3)
        return palette_img


# ============================================================================
# SIZE COMPARISON FUNCTION (FIXED)
# ============================================================================

def print_size_comparison(image_array, compressed, ratio):
    """Print actual size comparison with visual bar"""
    
    # Original size in bytes
    original_bytes = image_array.nbytes
    original_kb = original_bytes / 1024
    original_mb = original_kb / 1024
    
    # Get image dimensions
    h, w, c = image_array.shape
    
    # Get actual number of unique colors in compressed image
    compressed_flat = compressed.reshape(-1, c)
    k = len(np.unique(compressed_flat, axis=0))
    
    # Calculate compressed size accurately
    # Palette: k colors × c channels × 1 byte each
    palette_bytes = k * c
    
    # Indices: need log2(k) bits per pixel
    bits_per_index = np.ceil(np.log2(k))
    index_bytes = (h * w * bits_per_index) / 8
    
    # Total compressed bytes
    compressed_bytes = palette_bytes + index_bytes
    compressed_kb = compressed_bytes / 1024
    compressed_mb = compressed_kb / 1024
    
    print("\n" + "="*70)
    print("📊 ACTUAL SIZE COMPARISON")
    print("="*70)
    
    # Create visual bar (40 characters max)
    max_bar = 40
    original_bar = max_bar
    compressed_bar = int(max_bar * (compressed_bytes / original_bytes))
    if compressed_bar < 1:
        compressed_bar = 1
    
    print(f"Original:   [{'█' * original_bar}] {original_kb:.2f} KB ({original_mb:.4f} MB)")
    print(f"             ↓")
    print(f"Compressed: [{'█' * compressed_bar}{'░' * (max_bar - compressed_bar)}] {compressed_kb:.2f} KB ({compressed_mb:.4f} MB)")
    print(f"\n📈 Results:")
    print(f"   • Compression ratio: {ratio:.3f}:1")
    print(f"   • Space saved: {original_bytes - compressed_bytes:.0f} bytes")
    print(f"   • Smaller by: {(1 - compressed_bytes/original_bytes)*100:.1f}%")
    print(f"   • Colors: {len(np.unique(image_array.reshape(-1, c), axis=0))} → {k}")
    
    # Simple explanation
    print(f"\n💡 Explanation:")
    print(f"   Original had {original_bytes:,} bytes of data")
    print(f"   After K-Means, we store:")
    print(f"   • {k} colors × {c} channels = {palette_bytes} bytes (palette)")
    print(f"   • {h*w:,} pixels × {bits_per_index:.0f} bits = {index_bytes:.0f} bytes (indices)")
    print(f"   Total: {palette_bytes} + {index_bytes:.0f} = {compressed_bytes:.0f} bytes")


# ============================================================================
# DEMONSTRATION WITH SIMPLE 1D EXAMPLE
# ============================================================================

def demonstrate_1d_kmeans():
    """Simple 1D example to understand K-Means"""
    
    print("\n" + "="*70)
    print("📝 SIMPLE 1D K-MEANS EXAMPLE")
    print("="*70)
    
    # Generate 1D data with 3 natural clusters
    np.random.seed(42)
    data = np.concatenate([
        np.random.normal(2, 0.5, 30),
        np.random.normal(5, 0.5, 30),
        np.random.normal(8, 0.5, 30)
    ]).reshape(-1, 1)
    
    print(f"\nGenerated {len(data)} data points with 3 natural clusters")
    
    # Run K-Means
    kmeans = KMeans(k=3, max_iters=50)
    kmeans.fit(data)
    
    # Visualize
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    
    # Original data
    axes[0].hist(data, bins=20, color='gray', alpha=0.7, edgecolor='black')
    axes[0].set_title('Step 0: Original Data\n(Unlabeled)', fontweight='bold')
    axes[0].set_xlabel('Value')
    axes[0].set_ylabel('Frequency')
    
    # Clustered result
    colors = ['red', 'green', 'blue']
    for i in range(3):
        cluster_data = data[kmeans.labels == i]
        axes[1].hist(cluster_data, bins=15, color=colors[i], alpha=0.5, 
                    label=f'Cluster {i+1}', edgecolor='black')
        # Mark center
        axes[1].axvline(x=kmeans.centers[i, 0], color=colors[i], 
                       linestyle='--', linewidth=2, label=f'Center {i+1}')
    
    axes[1].set_title('Step 3-5: Final Clusters & Centers', fontweight='bold')
    axes[1].set_xlabel('Value')
    axes[1].legend()
    
    # Convergence history
    history = np.array([center[0,0] for center in kmeans.history])
    iterations = range(len(history))
    
    axes[2].plot(iterations, history, 'ro-', linewidth=2, markersize=6)
    axes[2].set_title('Step 6: Center Movement\n(Convergence)', fontweight='bold')
    axes[2].set_xlabel('Iteration')
    axes[2].set_ylabel('Center Value')
    axes[2].grid(True, alpha=0.3)
    
    plt.suptitle('K-Means Algorithm Steps Visualization', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Print results
    print(f"\n📊 Results:")
    print(f"   Final centers: {kmeans.centers.flatten()}")
    print(f"   Objective J(V): {kmeans.J:.4f}")
    print(f"   Iterations: {len(kmeans.history)-1}")


# ============================================================================
# IMAGE COMPRESSION DEMONSTRATION (WITH SIZE COMPARISON)
# ============================================================================

def demonstrate_image_compression(image_path="Lenna.png"):
    """Demonstrate K-Means for image compression with size comparison"""
    
    # Load image
    if os.path.exists(image_path):
        img = Image.open(image_path).convert('RGB')
        img_array = np.array(img)
        # Resize for faster processing
        img_array = np.array(Image.fromarray(img_array).resize((256, 256)))
        print(f"✅ Loaded: {image_path}")
    else:
        # Create test image
        print("⚠️ Creating test pattern")
        img_array = np.zeros((200, 200, 3), dtype=np.uint8)
        # Gradient
        for i in range(200):
            img_array[i, :, 0] = i
            img_array[i, :, 1] = 255 - i
            img_array[i, :, 2] = 128
        # Add solid color region
        img_array[50:150, 50:150] = [255, 0, 0]
    
    print(f"\n📊 Image info:")
    print(f"   Shape: {img_array.shape}")
    print(f"   Total pixels: {img_array.shape[0] * img_array.shape[1]}")
    print(f"   Unique colors: {len(np.unique(img_array.reshape(-1, 3), axis=0))}")
    
    # Test different K values (only 5 values to fit in 2x3 grid)
    k_values = [2, 4, 8, 16, 32]  # 5 values, plus original = 6 plots (2x3)
    results = {}
    
    print("\n" + "="*70)
    print("🔷 K-MEANS IMAGE COMPRESSION")
    print("="*70)
    
    # Create 2x3 grid
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    
    # Flatten axes for easier indexing
    axes_flat = axes.flatten()
    
    # Original image at position 0
    axes_flat[0].imshow(img_array)
    axes_flat[0].set_title('Original Image', fontweight='bold')
    axes_flat[0].axis('off')
    
    # Compress with different K (positions 1-5)
    for idx, k in enumerate(k_values):
        pos = idx + 1  # Start from position 1
        
        print(f"\n🔹 Testing K = {k}...")
        
        compressor = KMeansImageCompression(k=k)
        compressed, ratio, labels = compressor.compress(img_array)
        results[k] = {'compressed': compressed, 'ratio': ratio}
        
        axes_flat[pos].imshow(compressed)
        axes_flat[pos].set_title(f'K = {k}\nRatio: {ratio:.2f}:1', fontweight='bold')
        axes_flat[pos].axis('off')
    
    plt.suptitle('K-Means Image Compression with Different K Values', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Detailed analysis for K=16
    print("\n" + "="*70)
    print("🔍 DETAILED ANALYSIS (K=16)")
    print("="*70)
    
    compressor = KMeansImageCompression(k=16)
    compressed, ratio, labels = compressor.compress(img_array)
    
    # ===== ADD SIZE COMPARISON HERE =====
    print_size_comparison(img_array, compressed, ratio)
    # ====================================
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    
    # Original
    axes[0, 0].imshow(img_array)
    axes[0, 0].set_title('Original Image', fontweight='bold')
    axes[0, 0].axis('off')
    
    # Compressed
    axes[0, 1].imshow(compressed)
    axes[0, 1].set_title(f'Compressed (K=16)\nRatio: {ratio:.2f}:1', fontweight='bold')
    axes[0, 1].axis('off')
    
    # Difference
    diff = np.mean((img_array.astype(float) - compressed.astype(float)) ** 2, axis=2)
    im = axes[0, 2].imshow(diff, cmap='hot')
    axes[0, 2].set_title('Error Map\n(Bright = More Error)', fontweight='bold')
    axes[0, 2].axis('off')
    plt.colorbar(im, ax=axes[0, 2])
    
    # Color palette
    palette = compressor.palette
    palette_img = palette.reshape(1, -1, 3)
    axes[1, 0].imshow(palette_img, aspect='auto')
    axes[1, 0].set_title(f'Color Palette\n({len(palette)} colors)', fontweight='bold')
    axes[1, 0].set_yticks([])
    axes[1, 0].set_xlabel('Color Index')
    
    # Cluster distribution
    unique, counts = np.unique(labels, return_counts=True)
    axes[1, 1].bar(unique, counts, color=palette[unique]/255, edgecolor='black')
    axes[1, 1].set_title('Cluster Distribution', fontweight='bold')
    axes[1, 1].set_xlabel('Cluster Index')
    axes[1, 1].set_ylabel('Pixel Count')
    axes[1, 1].grid(True, alpha=0.3)
    
    # Statistics
    axes[1, 2].axis('off')
    orig_unique = len(np.unique(img_array.reshape(-1, 3), axis=0))
    stats = f"""📊 COMPRESSION STATISTICS

Original:
• Colors: {orig_unique}
• Size: {img_array.nbytes/1024:.1f} KB

Compressed (K=16):
• Colors: 16
• Palette: {16*3*8} bits
• Indices: {labels.size * 4} bits
• Total: {(16*3*8 + labels.size*4)/8/1024:.1f} KB
• Ratio: {ratio:.2f}:1

K-Means Info:
• Objective J: {compressor.kmeans.J:.2f}
• Iterations: {len(compressor.kmeans.history)-1}"""
    
    axes[1, 2].text(0.1, 0.5, stats, fontsize=10, fontfamily='monospace',
                   bbox=dict(boxstyle="round,pad=0.5", facecolor='#e6f3ff', edgecolor='blue'))
    
    plt.suptitle('K-Means Image Compression - Detailed Analysis', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    return results


# ============================================================================
# STEP-BY-STEP VISUALIZATION OF THE ALGORITHM
# ============================================================================

def visualize_algorithm_steps():
    """Visualize each step of K-Means algorithm"""
    
    print("\n" + "="*70)
    print("🔄 K-MEANS ALGORITHM STEPS VISUALIZATION")
    print("="*70)
    
    # Generate 2D data for easy visualization
    np.random.seed(42)
    data = np.concatenate([
        np.random.normal([2, 2], 0.5, (20, 2)),
        np.random.normal([5, 5], 0.5, (20, 2)),
        np.random.normal([8, 2], 0.5, (20, 2))
    ])
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    
    # Step 1: Random centers
    kmeans = KMeans(k=3, max_iters=1)
    kmeans.centers = np.array([[3, 3], [4, 4], [6, 3]])  # Initial centers
    kmeans.history = [kmeans.centers.copy()]
    
    axes[0, 0].scatter(data[:, 0], data[:, 1], c='gray', alpha=0.5)
    axes[0, 0].scatter(kmeans.centers[:, 0], kmeans.centers[:, 1], 
                      c=['red', 'green', 'blue'], s=200, marker='X', edgecolor='black')
    axes[0, 0].set_title('Step 1: Randomly select c centers', fontweight='bold')
    
    # Step 2-3: Assign based on distance
    distances = kmeans.calculate_distances(data, kmeans.centers)
    labels = kmeans.assign_clusters(distances)
    
    colors = ['red', 'green', 'blue']
    for i in range(3):
        cluster_data = data[labels == i]
        axes[0, 1].scatter(cluster_data[:, 0], cluster_data[:, 1], 
                          c=colors[i], alpha=0.5, label=f'Cluster {i+1}')
    axes[0, 1].scatter(kmeans.centers[:, 0], kmeans.centers[:, 1], 
                      c=['red', 'green', 'blue'], s=200, marker='X', edgecolor='black')
    axes[0, 1].set_title('Steps 2-3: Assign points to nearest center', fontweight='bold')
    
    # Step 4: Update centers
    new_centers = kmeans.update_centers(data, labels)
    
    for i in range(3):
        cluster_data = data[labels == i]
        axes[0, 2].scatter(cluster_data[:, 0], cluster_data[:, 1], 
                          c=colors[i], alpha=0.3)
    axes[0, 2].scatter(kmeans.centers[:, 0], kmeans.centers[:, 1], 
                      c=['red', 'green', 'blue'], s=200, marker='X', edgecolor='black', alpha=0.3)
    axes[0, 2].scatter(new_centers[:, 0], new_centers[:, 1], 
                      c=['red', 'green', 'blue'], s=200, marker='o', edgecolor='black')
    
    for i in range(3):
        axes[0, 2].arrow(kmeans.centers[i, 0], kmeans.centers[i, 1],
                        new_centers[i, 0] - kmeans.centers[i, 0],
                        new_centers[i, 1] - kmeans.centers[i, 1],
                        head_width=0.1, color=colors[i])
    
    axes[0, 2].set_title('Step 4: Recalculate centers (equation 8)', fontweight='bold')
    
    # Step 5: Recalculate distances
    kmeans.centers = new_centers
    distances = kmeans.calculate_distances(data, kmeans.centers)
    new_labels = kmeans.assign_clusters(distances)
    
    for i in range(3):
        cluster_data = data[new_labels == i]
        axes[1, 0].scatter(cluster_data[:, 0], cluster_data[:, 1], 
                          c=colors[i], alpha=0.5)
    axes[1, 0].scatter(kmeans.centers[:, 0], kmeans.centers[:, 1], 
                      c=['red', 'green', 'blue'], s=200, marker='X', edgecolor='black')
    axes[1, 0].set_title('Step 5: Recalculate distances', fontweight='bold')
    
    # Step 6: Check convergence
    moved = np.sum(labels != new_labels)
    
    axes[1, 1].scatter(data[:, 0], data[:, 1], c='gray', alpha=0.5)
    axes[1, 1].scatter(kmeans.centers[:, 0], kmeans.centers[:, 1], 
                      c=['red', 'green', 'blue'], s=200, marker='X', edgecolor='black')
    axes[1, 1].set_title(f'Step 6: Check if any point moved\n{moved} points moved', fontweight='bold')
    
    # Final result
    for i in range(3):
        cluster_data = data[new_labels == i]
        axes[1, 2].scatter(cluster_data[:, 0], cluster_data[:, 1], 
                          c=colors[i], alpha=0.5)
    axes[1, 2].scatter(kmeans.centers[:, 0], kmeans.centers[:, 1], 
                      c=['red', 'green', 'blue'], s=200, marker='X', edgecolor='black')
    axes[1, 2].set_title('Final Clusters (Repeat until no movement)', fontweight='bold')
    
    plt.suptitle('K-Means Algorithm Step-by-Step', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()


# ============================================================================
# MAIN EXECUTION
# ============================================================================

if __name__ == "__main__":
    print("="*70)
    print("🔷 K-MEANS CLUSTERING ALGORITHM")
    print("="*70)
    print("\n📚 Reference: McQueen (1967) - Unsupervised Machine Learning")
    
    # Show algorithm steps
    visualize_algorithm_steps()
    
    # Simple 1D example
    demonstrate_1d_kmeans()
    
    # Image compression demo
    results = demonstrate_image_compression("Lenna.png")    
    results1 = demonstrate_image_compression("camera.png")
```
