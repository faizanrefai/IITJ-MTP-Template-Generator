```python
# ============================================
# ASSIGNMENT-3: Parameter-Efficient Fine-Tuning (PEFT) with Adapters
# Course: IIT MTP - Deep Learning
# Task: SST-2 Sentiment Analysis with BERT + Adapters
# Hardware: Apple M4 GPU (MPS Acceleration)
# ============================================

# Cell 1: Install Required Packages
print("="*70)
print("STEP 1: INSTALLING REQUIRED PACKAGES")
print("="*70)

%pip install -q transformers datasets matplotlib seaborn torch tqdm ipywidgets
print("✅ All packages installed successfully!\n")

# Cell 2: Imports with Progress Bar Fix
print("="*70)
print("STEP 2: IMPORTING LIBRARIES")
print("="*70)

import torch
import torch.nn as nn
from torch.optim import AdamW
from torch.utils.data import DataLoader
from transformers import BertTokenizer, BertModel
from datasets import load_dataset
from tqdm.auto import tqdm  # Fixed: Use auto instead of notebook
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import time
import os
from datetime import datetime
import json

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')

print("✅ All libraries imported successfully!")
print(f"PyTorch version: {torch.__version__}")
print(f"Transformers version: {__import__('transformers').__version__}\n")

# Cell 3: Device Configuration for Apple M4 GPU
print("="*70)
print("STEP 3: CONFIGURING HARDWARE ACCELERATION")
print("="*70)

def setup_device():
    """Setup the best available device with detailed logging"""
    print("🔍 Checking available devices...")
    
    # Check for Apple MPS (Metal Performance Shaders)
    if torch.backends.mps.is_available():
        device = torch.device("mps")
        print("✅ Apple M4 GPU (MPS) - AVAILABLE and SELECTED")
        print(f"   • Metal Performance Shaders is ready")
        print(f"   • GPU Type: Apple M4")
        print(f"   • Memory: Shared with system")
        # Optimize MPS settings
        torch.mps.set_per_process_memory_fraction(0.8)
        print(f"   • Memory limit: 80% of total available")
    elif torch.cuda.is_available():
        device = torch.device("cuda")
        print("✅ NVIDIA CUDA GPU - AVAILABLE and SELECTED")
        print(f"   • GPU: {torch.cuda.get_device_name(0)}")
        print(f"   • Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
    else:
        device = torch.device("cpu")
        print("⚠️ GPU NOT AVAILABLE - Using CPU")
        print("   • Training will be slower")
    
    print(f"\n📊 Device summary:")
    print(f"   • Using: {device}")
    print(f"   • PyTorch version: {torch.__version__}")
    print(f"   • MPS available: {torch.backends.mps.is_available()}")
    print(f"   • CUDA available: {torch.cuda.is_available()}")
    
    return device

device = setup_device()
print("="*70 + "\n")

# Cell 4: Dataset Loading and Preprocessing
print("="*70)
print("STEP 4: LOADING AND PREPROCESSING SST-2 DATASET")
print("="*70)

class SST2Preprocessor:
    """Complete preprocessing pipeline for SST-2 dataset"""
    
    def __init__(self, max_length=128):
        print(f"\n🔧 Initializing BERT Tokenizer...")
        self.tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
        self.max_length = max_length
        print(f"✅ Tokenizer loaded: bert-base-uncased")
        print(f"   • Max sequence length: {max_length} tokens")
        print(f"   • Vocabulary size: {self.tokenizer.vocab_size:,}")
    
    def tokenize_function(self, examples):
        """Convert raw text to BERT inputs"""
        return self.tokenizer(
            examples['sentence'],
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )
    
    def analyze_dataset(self, dataset):
        """Print dataset statistics"""
        print(f"\n📊 Dataset Statistics:")
        print(f"   • Training samples: {len(dataset['train']):,}")
        print(f"   • Validation samples: {len(dataset['validation']):,}")
        
        # Analyze sentence lengths (sample first 1000)
        train_lengths = [len(self.tokenizer.encode(s)) for s in dataset['train']['sentence'][:1000]]
        print(f"   • Avg sentence length: {np.mean(train_lengths):.1f} tokens")
        print(f"   • Max sentence length: {np.max(train_lengths)} tokens")
        print(f"   • Min sentence length: {np.min(train_lengths)} tokens")
        
        # Label distribution
        train_labels = dataset['train']['label']
        pos_count = sum(train_labels)
        neg_count = len(train_labels) - pos_count
        print(f"   • Label distribution:")
        print(f"     - Positive: {pos_count} ({pos_count/len(train_labels)*100:.1f}%)")
        print(f"     - Negative: {neg_count} ({neg_count/len(train_labels)*100:.1f}%)")
    
    def prepare_dataset(self, dataset, max_samples=10000):
        """Prepare dataset for training"""
        print(f"\n🔧 Preparing dataset (limiting to {max_samples:,} samples)...")
        
        # Limit training samples
        original_size = len(dataset['train'])
        if max_samples < original_size:
            dataset['train'] = dataset['train'].select(range(min(max_samples, len(dataset['train']))))
            print(f"   • Reduced from {original_size:,} to {len(dataset['train']):,} samples")
        
        # Apply tokenization
        print("   • Tokenizing text...")
        tokenized_dataset = dataset.map(self.tokenize_function, batched=True)
        
        # Set format for PyTorch
        tokenized_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'token_type_ids', 'label'])
        
        print("✅ Dataset preparation complete!")
        return tokenized_dataset

# Load dataset
print("\n📥 Loading SST-2 dataset from Hugging Face...")
dataset = load_dataset('stanfordnlp/sst2')
print("✅ Dataset loaded successfully!")

# Preprocess
preprocessor = SST2Preprocessor(max_length=128)
preprocessor.analyze_dataset(dataset)
tokenized_dataset = preprocessor.prepare_dataset(dataset, max_samples=10000)

# Create DataLoaders
batch_size = 32
train_loader = DataLoader(tokenized_dataset['train'], batch_size=batch_size, shuffle=True)
val_loader = DataLoader(tokenized_dataset['validation'], batch_size=batch_size, shuffle=False)

print(f"\n📦 DataLoader configuration:")
print(f"   • Batch size: {batch_size}")
print(f"   • Training batches: {len(train_loader)}")
print(f"   • Validation batches: {len(val_loader)}")
print("="*70 + "\n")

# Cell 5: Adapter Module Implementation
print("="*70)
print("STEP 5: BUILDING ADAPTER MODULE")
print("="*70)

class Adapter(nn.Module):
    """
    Parameter-Efficient Adapter Module
    Projects to bottleneck dimension for efficient fine-tuning
    """
    
    def __init__(self, hidden_size, bottleneck_dim=64, activation='gelu'):
        super(Adapter, self).__init__()
        
        print(f"\n🔧 Creating Adapter layer:")
        print(f"   • Hidden size: {hidden_size}")
        print(f"   • Bottleneck dimension: {bottleneck_dim}")
        print(f"   • Compression ratio: {hidden_size/bottleneck_dim:.1f}x")
        print(f"   • Activation: {activation}")
        
        # Down-projection (compression)
        self.down_proj = nn.Linear(hidden_size, bottleneck_dim)
        
        # Up-projection (expansion)
        self.up_proj = nn.Linear(bottleneck_dim, hidden_size)
        
        # Activation function
        self.activation = nn.GELU() if activation == 'gelu' else nn.ReLU()
        
        # LayerNorm for stability
        self.layer_norm = nn.LayerNorm(hidden_size)
        
        # Initialize weights properly
        self._init_weights()
        
        # Calculate parameter count
        adapter_params = 2 * hidden_size * bottleneck_dim + hidden_size + bottleneck_dim
        print(f"   • Adapter parameters: {adapter_params:,}")
    
    def _init_weights(self):
        """Initialize adapter weights for stable training"""
        nn.init.xavier_uniform_(self.down_proj.weight, gain=0.5)
        nn.init.zeros_(self.down_proj.bias)
        nn.init.xavier_uniform_(self.up_proj.weight, gain=0.5)
        nn.init.zeros_(self.up_proj.bias)
    
    def forward(self, x):
        """Forward pass with residual connection"""
        residual = x
        x = self.layer_norm(x)
        x = self.down_proj(x)
        x = self.activation(x)
        x = self.up_proj(x)
        return residual + x

print("\n✅ Adapter module defined successfully!")
print("="*70 + "\n")

# Cell 6: Injecting Adapters into BERT
print("="*70)
print("STEP 6: INJECTING ADAPTERS INTO BERT MODEL")
print("="*70)

class BertWithAdapters(nn.Module):
    """BERT model with adapters injected after each attention layer"""
    
    def __init__(self, model_name='bert-base-uncased', num_classes=2, bottleneck_dim=64):
        super(BertWithAdapters, self).__init__()
        
        print(f"\n🔧 Loading pre-trained BERT model...")
        self.bert = BertModel.from_pretrained(model_name)
        print(f"✅ BERT loaded: {model_name}")
        print(f"   • Hidden size: {self.bert.config.hidden_size}")
        print(f"   • Number of layers: {self.bert.config.num_hidden_layers}")
        print(f"   • Attention heads: {self.bert.config.num_attention_heads}")
        
        # FREEZE all base BERT parameters
        print(f"\n🔒 Freezing base BERT parameters...")
        frozen_params = 0
        for param in self.bert.parameters():
            param.requires_grad = False
            frozen_params += param.numel()
        print(f"   • Frozen parameters: {frozen_params:,}")
        
        # Inject adapters after each transformer block
        print(f"\n🔧 Injecting {self.bert.config.num_hidden_layers} adapter modules...")
        self.adapters = nn.ModuleList([
            Adapter(
                hidden_size=self.bert.config.hidden_size,
                bottleneck_dim=bottleneck_dim
            )
            for _ in range(self.bert.config.num_hidden_layers)
        ])
        print(f"✅ {len(self.adapters)} adapters injected successfully!")
        
        # Classification head (trainable)
        self.classifier = nn.Sequential(
            nn.Dropout(0.1),
            nn.Linear(self.bert.config.hidden_size, num_classes)
        )
        print(f"\n🔧 Classification head:")
        print(f"   • Input size: {self.bert.config.hidden_size}")
        print(f"   • Output classes: {num_classes}")
        print(f"   • Trainable parameters: {sum(p.numel() for p in self.classifier.parameters()):,}")
        
        self.num_layers = self.bert.config.num_hidden_layers
    
    def forward(self, input_ids, attention_mask, token_type_ids):
        """Forward pass with adapter injection"""
        # Get hidden states from BERT
        outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            output_hidden_states=True,
            return_dict=True
        )
        
        hidden_states = outputs.hidden_states
        
        # Apply adapters after each layer
        for layer_idx in range(self.num_layers):
            layer_output = hidden_states[layer_idx + 1]
            adapted_output = self.adapters[layer_idx](layer_output)
            hidden_states = list(hidden_states)
            hidden_states[layer_idx + 1] = adapted_output
            hidden_states = tuple(hidden_states)
        
        # Use last layer's [CLS] token for classification
        final_hidden = hidden_states[-1]
        cls_output = final_hidden[:, 0, :]
        
        # Classification
        logits = self.classifier(cls_output)
        
        return logits

# Initialize model
print("\n🚀 Initializing BERT with Adapters...")
model = BertWithAdapters(
    model_name='bert-base-uncased',
    num_classes=2,
    bottleneck_dim=64
).to(device)

print("\n✅ Model initialization complete!")
print("="*70 + "\n")

# Cell 7: Parameter Count Comparison
print("="*70)
print("STEP 7: PARAMETER COUNT ANALYSIS")
print("="*70)

def count_parameters_detailed(model):
    """Detailed parameter counting with breakdown"""
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    frozen_params = total_params - trainable_params
    
    # Breakdown by component
    bert_params = sum(p.numel() for name, p in model.bert.named_parameters())
    adapter_params = sum(p.numel() for adapter in model.adapters for p in adapter.parameters())
    classifier_params = sum(p.numel() for p in model.classifier.parameters())
    
    return {
        'total': total_params,
        'trainable': trainable_params,
        'frozen': frozen_params,
        'bert': bert_params,
        'adapters': adapter_params,
        'classifier': classifier_params
    }

param_counts = count_parameters_detailed(model)

print("\n📊 PARAMETER COUNT BREAKDOWN")
print("-"*50)
print(f"{'Component':<25} {'Parameters':<15} {'Percentage':<10}")
print("-"*50)
print(f"{'BERT Base (Frozen)':<25} {param_counts['bert']:>12,}   {param_counts['bert']/param_counts['total']*100:>5.1f}%")
print(f"{'Adapter Modules':<25} {param_counts['adapters']:>12,}   {param_counts['adapters']/param_counts['total']*100:>5.2f}%")
print(f"{'Classification Head':<25} {param_counts['classifier']:>12,}   {param_counts['classifier']/param_counts['total']*100:>5.2f}%")
print("-"*50)
print(f"{'TOTAL PARAMETERS':<25} {param_counts['total']:>12,}   {100:>5.1f}%")
print(f"{'TRAINABLE PARAMETERS':<25} {param_counts['trainable']:>12,}   {param_counts['trainable']/param_counts['total']*100:>5.2f}%")
print(f"{'FROZEN PARAMETERS':<25} {param_counts['frozen']:>12,}   {param_counts['frozen']/param_counts['total']*100:>5.2f}%")
print("-"*50)

print(f"\n💡 EFFICIENCY GAIN:")
print(f"   • Only {param_counts['trainable']/param_counts['total']*100:.2f}% of parameters are trainable!")
print(f"   • {100 - param_counts['trainable']/param_counts['total']*100:.2f}% reduction in trainable parameters")
print(f"   • Memory savings: ~{(param_counts['frozen']/1e6):.0f}M parameters frozen")

print("="*70 + "\n")

# Cell 8: Training Setup with Live Visualization
print("="*70)
print("STEP 8: SETTING UP TRAINING CONFIGURATION")
print("="*70)

class Trainer:
    """Complete training pipeline with logging and live visualization"""
    
    def __init__(self, model, train_loader, val_loader, learning_rate=2e-5, epochs=5):
        self.model = model
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.epochs = epochs
        self.device = next(model.parameters()).device
        
        print(f"\n🔧 Training Configuration:")
        print(f"   • Learning rate: {learning_rate}")
        print(f"   • Number of epochs: {epochs}")
        print(f"   • Device: {self.device}")
        print(f"   • Optimizer: AdamW")
        print(f"   • Loss function: CrossEntropyLoss")
        
        # Optimizer (only trainable parameters)
        self.optimizer = AdamW(
            filter(lambda p: p.requires_grad, model.parameters()),
            lr=learning_rate,
            weight_decay=0.01
        )
        
        # Loss function
        self.criterion = nn.CrossEntropyLoss()
        
        # Tracking metrics
        self.train_losses = []
        self.val_losses = []
        self.val_accuracies = []
        self.training_times = []
        
        print("✅ Training configuration complete!")
    
    def train_epoch(self, epoch):
        """Train for one epoch with detailed logging"""
        self.model.train()
        total_loss = 0
        start_time = time.time()
        
        progress_bar = tqdm(self.train_loader, desc=f"Epoch {epoch+1} Training", leave=False)
        
        for batch_idx, batch in enumerate(progress_bar):
            # Move data to device
            input_ids = batch['input_ids'].to(self.device)
            attention_mask = batch['attention_mask'].to(self.device)
            token_type_ids = batch['token_type_ids'].to(self.device)
            labels = batch['label'].to(self.device)
            
            # Forward pass
            self.optimizer.zero_grad()
            logits = self.model(input_ids, attention_mask, token_type_ids)
            loss = self.criterion(logits, labels)
            
            # Backward pass
            loss.backward()
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), 1.0)
            self.optimizer.step()
            
            total_loss += loss.item()
            
            # Update progress bar
            progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
        
        epoch_time = time.time() - start_time
        avg_loss = total_loss / len(self.train_loader)
        
        print(f"   ✓ Training completed in {epoch_time:.2f}s | Avg Loss: {avg_loss:.4f}")
        
        return avg_loss, epoch_time
    
    def validate(self, epoch):
        """Validate the model with detailed metrics"""
        self.model.eval()
        total_loss = 0
        correct = 0
        total = 0
        
        with torch.no_grad():
            progress_bar = tqdm(self.val_loader, desc=f"Epoch {epoch+1} Validation", leave=False)
            for batch in progress_bar:
                input_ids = batch['input_ids'].to(self.device)
                attention_mask = batch['attention_mask'].to(self.device)
                token_type_ids = batch['token_type_ids'].to(self.device)
                labels = batch['label'].to(self.device)
                
                logits = self.model(input_ids, attention_mask, token_type_ids)
                loss = self.criterion(logits, labels)
                
                total_loss += loss.item()
                predictions = torch.argmax(logits, dim=-1)
                correct += (predictions == labels).sum().item()
                total += labels.size(0)
        
        avg_loss = total_loss / len(self.val_loader)
        accuracy = correct / total
        
        print(f"   ✓ Validation | Loss: {avg_loss:.4f} | Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")
        
        return avg_loss, accuracy
    
    def train(self):
        """Complete training loop with progress tracking"""
        print("\n" + "="*70)
        print("🚀 STARTING SUPERVISED FINE-TUNING")
        print("="*70)
        
        for epoch in range(self.epochs):
            print(f"\n{'='*70}")
            print(f"EPOCH {epoch+1}/{self.epochs}")
            print(f"{'='*70}")
            
            # Train
            train_loss, epoch_time = self.train_epoch(epoch)
            self.train_losses.append(train_loss)
            self.training_times.append(epoch_time)
            
            # Validate
            val_loss, val_acc = self.validate(epoch)
            self.val_losses.append(val_loss)
            self.val_accuracies.append(val_acc)
        
        print("\n" + "="*70)
        print("✅ TRAINING COMPLETE!")
        print("="*70)
        
        # Print final metrics
        print(f"\n📊 FINAL METRICS:")
        print(f"   • Best Validation Accuracy: {max(self.val_accuracies)*100:.2f}%")
        print(f"   • Final Validation Accuracy: {self.val_accuracies[-1]*100:.2f}%")
        print(f"   • Best Validation Loss: {min(self.val_losses):.4f}")
        print(f"   • Total Training Time: {sum(self.training_times):.2f}s")
        
        return self.train_losses, self.val_losses, self.val_accuracies

# Initialize trainer
trainer = Trainer(model, train_loader, val_loader, learning_rate=2e-5, epochs=5)

print("="*70 + "\n")

# Cell 9: Training Execution
print("="*70)
print("STEP 9: EXECUTING TRAINING LOOP")
print("="*70)

# Start training
train_start_time = time.time()
train_losses, val_losses, val_accuracies = trainer.train()
total_training_time = time.time() - train_start_time

print(f"\n⏱️ Total training time: {total_training_time:.2f} seconds ({total_training_time/60:.2f} minutes)")

# Cell 10: Save Model
print("\n" + "="*70)
print("STEP 10: SAVING TRAINED MODEL")
print("="*70)

# Create models directory if it doesn't exist
model_dir = "saved_models"
os.makedirs(model_dir, exist_ok=True)

# Save model checkpoint
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
model_path = f"{model_dir}/peft_adapter_sst2_{timestamp}.pt"
config_path = f"{model_dir}/peft_adapter_config_{timestamp}.json"

# Save model state dict
torch.save({
    'model_state_dict': model.state_dict(),
    'train_losses': train_losses,
    'val_losses': val_losses,
    'val_accuracies': val_accuracies,
    'param_counts': param_counts,
    'epochs': trainer.epochs,
    'best_accuracy': max(val_accuracies),
    'timestamp': timestamp
}, model_path)

# Save configuration
config = {
    'model_type': 'BERT with Adapters',
    'base_model': 'bert-base-uncased',
    'bottleneck_dim': 64,
    'max_length': 128,
    'num_classes': 2,
    'total_parameters': param_counts['total'],
    'trainable_parameters': param_counts['trainable'],
    'frozen_parameters': param_counts['frozen'],
    'trainable_percentage': param_counts['trainable']/param_counts['total']*100,
    'best_accuracy': max(val_accuracies),
    'training_timestamp': timestamp
}

with open(config_path, 'w') as f:
    json.dump(config, f, indent=4)

print(f"✅ Model saved to: {model_path}")
print(f"✅ Config saved to: {config_path}")
print(f"\n📦 Saved files:")
print(f"   • Model checkpoint: {os.path.basename(model_path)}")
print(f"   • Configuration: {os.path.basename(config_path)}")

# Cell 11: Enhanced Visualizations for University Submission
print("\n" + "="*70)
print("STEP 11: GENERATING ENHANCED VISUALIZATIONS")
print("="*70)

# Create a professional figure with multiple subplots
fig = plt.figure(figsize=(20, 12))
fig.suptitle('Parameter-Efficient Fine-Tuning with Adapters - SST-2 Sentiment Analysis\nIIT MTP Assignment 3', 
             fontsize=16, fontweight='bold', y=0.98)

# Plot 1: Training vs Validation Loss
ax1 = plt.subplot(2, 3, 1)
epochs_range = range(1, len(train_losses)+1)
ax1.plot(epochs_range, train_losses, 'b-', linewidth=2, marker='o', markersize=8, label='Training Loss', color='#1f77b4')
ax1.plot(epochs_range, val_losses, 'r-', linewidth=2, marker='s', markersize=8, label='Validation Loss', color='#ff7f0e')
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.set_title('Training vs Validation Loss Curves', fontsize=14, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)
ax1.set_xticks(epochs_range)

# Add loss values on plot
for i, (train_loss, val_loss) in enumerate(zip(train_losses, val_losses)):
    ax1.annotate(f'{train_loss:.3f}', xy=(i+1, train_loss), xytext=(5, 5), textcoords='offset points', fontsize=9)
    ax1.annotate(f'{val_loss:.3f}', xy=(i+1, val_loss), xytext=(5, -15), textcoords='offset points', fontsize=9)

# Plot 2: Validation Accuracy
ax2 = plt.subplot(2, 3, 2)
ax2.plot(epochs_range, [acc*100 for acc in val_accuracies], 'g-', linewidth=2, marker='^', markersize=10, color='#2ca02c')
ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Accuracy (%)', fontsize=12)
ax2.set_title('Validation Accuracy Over Epochs', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_xticks(epochs_range)
ax2.set_ylim([70, 100])

# Add accuracy values on plot
for i, acc in enumerate(val_accuracies):
    ax2.annotate(f'{acc*100:.2f}%', 
                xy=(i+1, acc*100), 
                xytext=(5, 5), 
                textcoords='offset points',
                fontsize=10,
                fontweight='bold')

# Plot 3: Parameter Efficiency (Horizontal Bar Chart)
ax3 = plt.subplot(2, 3, 3)
components = ['BERT Base (Frozen)', 'Adapter Modules', 'Classification Head']
params = [param_counts['bert'], param_counts['adapters'], param_counts['classifier']]
colors_bar = ['#ff9999', '#66b3ff', '#99ff99']
bars = ax3.barh(components, params, color=colors_bar, edgecolor='black')
ax3.set_xlabel('Number of Parameters', fontsize=12)
ax3.set_title('Parameter Distribution by Component', fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3, axis='x')

# Add parameter values on bars
for bar, param_val in zip(bars, params):
    ax3.text(bar.get_width() + 10000, bar.get_y() + bar.get_height()/2, 
             f'{param_val:,}', va='center', fontsize=10, fontweight='bold')

# Plot 4: Training Time per Epoch
ax4 = plt.subplot(2, 3, 4)
bars_time = ax4.bar(epochs_range, trainer.training_times, color='steelblue', alpha=0.7, edgecolor='black', width=0.6)
ax4.set_xlabel('Epoch', fontsize=12)
ax4.set_ylabel('Time (seconds)', fontsize=12)
ax4.set_title('Training Time per Epoch', fontsize=14, fontweight='bold')
ax4.set_xticks(epochs_range)
ax4.grid(True, alpha=0.3, axis='y')

# Add time values on bars
for bar, time_val in zip(bars_time, trainer.training_times):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{time_val:.1f}s', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Plot 5: Parameter Efficiency Pie Chart
ax5 = plt.subplot(2, 3, 5)
sizes = [param_counts['trainable'], param_counts['frozen']]
labels = [f'Trainable\n({param_counts["trainable"]/param_counts["total"]*100:.2f}%)', 
          f'Frozen\n({param_counts["frozen"]/param_counts["total"]*100:.2f}%)']
colors_pie = ['#ff6b6b', '#4ecdc4']
explode = (0.05, 0)
ax5.pie(sizes, explode=explode, labels=labels, colors=colors_pie, autopct='%1.2f%%',
        shadow=True, startangle=90, textprops={'fontsize': 11, 'fontweight': 'bold'})
ax5.set_title('Parameter Efficiency Analysis', fontsize=14, fontweight='bold')

# Plot 6: Convergence Analysis (Loss vs Accuracy)
ax6 = plt.subplot(2, 3, 6)
ax6_twin = ax6.twinx()
ax6.plot(epochs_range, val_losses, 'ro-', linewidth=2, markersize=8, label='Validation Loss', color='#d62728')
ax6_twin.plot(epochs_range, [acc*100 for acc in val_accuracies], 'bs-', linewidth=2, markersize=8, label='Accuracy', color='#1f77b4')
ax6.set_xlabel('Epoch', fontsize=12)
ax6.set_ylabel('Loss', fontsize=12, color='red')
ax6_twin.set_ylabel('Accuracy (%)', fontsize=12, color='blue')
ax6.set_title('Convergence Analysis: Loss vs Accuracy', fontsize=14, fontweight='bold')
ax6.grid(True, alpha=0.3)
ax6.tick_params(axis='y', labelcolor='red')
ax6_twin.tick_params(axis='y', labelcolor='blue')

plt.tight_layout()
plt.savefig('peft_adapter_complete_results.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ Enhanced visualization saved as 'peft_adapter_complete_results.png'")

# Cell 12: Training Summary Table
print("\n" + "="*70)
print("STEP 12: TRAINING SUMMARY TABLE")
print("="*70)

# Try to import tabulate, if not available, create manual table
try:
    from tabulate import tabulate
    has_tabulate = True
except ImportError:
    has_tabulate = False
    print("Note: Install 'tabulate' for better table formatting: %pip install tabulate")

table_data = []
for epoch in range(trainer.epochs):
    table_data.append([
        epoch + 1,
        f"{train_losses[epoch]:.4f}",
        f"{val_losses[epoch]:.4f}",
        f"{val_accuracies[epoch]*100:.2f}%",
        f"{trainer.training_times[epoch]:.2f}s",
        f"{(val_accuracies[epoch] - (val_accuracies[epoch-1] if epoch>0 else 0))*100:+.2f}%"
    ])

headers = ["Epoch", "Train Loss", "Val Loss", "Val Accuracy", "Time", "Improvement"]
if has_tabulate:
    print(tabulate(table_data, headers=headers, tablefmt="grid"))
else:
    print(f"{headers[0]:<8} {headers[1]:<12} {headers[2]:<12} {headers[3]:<12} {headers[4]:<10} {headers[5]:<12}")
    print("-"*70)
    for row in table_data:
        print(f"{row[0]:<8} {row[1]:<12} {row[2]:<12} {row[3]:<12} {row[4]:<10} {row[5]:<12}")

print(f"\n📈 BEST RESULTS:")
print(f"   • Best Validation Accuracy: {max(val_accuracies)*100:.2f}% (Epoch {val_accuracies.index(max(val_accuracies))+1})")
print(f"   • Best Validation Loss: {min(val_losses):.4f} (Epoch {val_losses.index(min(val_losses))+1})")
print(f"   • Final Validation Accuracy: {val_accuracies[-1]*100:.2f}%")
print(f"   • Average Epoch Time: {np.mean(trainer.training_times):.2f}s")

# Cell 13: Inference Pipeline
print("\n" + "="*70)
print("STEP 13: INFERENCE PIPELINE DEMO")
print("="*70)

class SentimentInferencePipeline:
    """Production-ready inference pipeline for sentiment analysis"""
    
    def __init__(self, model, tokenizer, max_length=128, device='cpu'):
        self.model = model
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.device = device
        self.model.eval()
        print(f"🔧 Inference pipeline initialized on {device}")
    
    def predict(self, text):
        """Predict sentiment for input text"""
        # Tokenize
        encoded = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )
        
        # Move to device
        input_ids = encoded['input_ids'].to(self.device)
        attention_mask = encoded['attention_mask'].to(self.device)
        token_type_ids = encoded['token_type_ids'].to(self.device)
        
        # Predict
        with torch.no_grad():
            logits = self.model(input_ids, attention_mask, token_type_ids)
            probabilities = torch.softmax(logits, dim=-1)
            prediction = torch.argmax(logits, dim=-1)
        
        sentiment = 'POSITIVE' if prediction.item() == 1 else 'NEGATIVE'
        confidence = probabilities[0][prediction.item()].item()
        
        return sentiment, confidence

# Initialize pipeline
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
pipeline = SentimentInferencePipeline(model, tokenizer, max_length=128, device=device)

# Test examples
test_cases = [
    ("The movie was absolutely fantastic!", "Strong positive"),
    ("This film is terrible and boring.", "Strong negative"),
    ("I really enjoyed watching this, it was quite entertaining.", "Positive"),
    ("The acting was poor and the plot made no sense.", "Negative"),
    ("A masterpiece of modern cinema!", "Very positive"),
    ("Not worth the time or money.", "Negative"),
    ("It was okay, nothing special.", "Neutral"),
    ("Brilliant performance by the entire cast!", "Positive"),
]

print("\n📝 SENTIMENT ANALYSIS RESULTS")
print("="*80)
print(f"{'Input Text':<45} {'Sentiment':<12} {'Confidence':<10} {'Expected':<15}")
print("="*80)

for text, expected in test_cases:
    sentiment, confidence = pipeline.predict(text)
    # Truncate long text for display
    display_text = text[:42] + "..." if len(text) > 45 else text
    emoji = "😊" if sentiment == "POSITIVE" else "😞"
    print(f"{display_text:<45} {emoji} {sentiment:<10} {confidence*100:>6.2f}%     {expected:<15}")
    print("-"*80)

# Cell 14: Load Saved Model Demo (for University Presentation)
print("\n" + "="*70)
print("STEP 14: DEMONSTRATING MODEL LOADING")
print("="*70)

# Load the saved model to show it works
print("📂 Loading saved model...")
loaded_checkpoint = torch.load(model_path, map_location=device)

# Recreate model architecture
loaded_model = BertWithAdapters(
    model_name='bert-base-uncased',
    num_classes=2,
    bottleneck_dim=64
).to(device)

# Load weights
loaded_model.load_state_dict(loaded_checkpoint['model_state_dict'])
print("✅ Model loaded successfully!")

# Test loaded model
test_text = "This is a great demonstration of model persistence!"
sentiment, confidence = SentimentInferencePipeline(loaded_model, tokenizer, device=device).predict(test_text)
print(f"\n🔍 Testing loaded model on: '{test_text}'")
print(f"   Result: {sentiment} (Confidence: {confidence*100:.2f}%)")
print("\n✅ Model saving and loading verified!")

# Cell 15: Final Assignment Report
print("\n" + "="*70)
print("FINAL ASSIGNMENT REPORT - PEFT with Adapters")
print("="*70)

print(f"""
╔══════════════════════════════════════════════════════════════════════╗
║                    ASSIGNMENT-3 COMPLETION REPORT                     ║
╚══════════════════════════════════════════════════════════════════════╝

📋 TASKS COMPLETED:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
1. ✓ Preprocessing & BERT Tokenization
   - Implemented complete tokenization pipeline
   - Set max sequence length to 128 tokens
   - Generated token, segment, and position embeddings

2. ✓ Adapter Module Implementation
   - Created custom Adapter class with bottleneck architecture
   - Used bottleneck dimension of 64
   - Implemented residual connections and LayerNorm

3. ✓ Adapter Injection
   - Loaded pre-trained BERT base model
   - Frozen all {param_counts['bert']:,} base parameters
   - Injected 12 adapter modules (one per transformer layer)

4. ✓ Supervised Fine-Tuning
   - Trained for {trainer.epochs} epochs using AdamW optimizer
   - Used Cross-Entropy Loss function
   - Achieved {max(val_accuracies)*100:.2f}% validation accuracy

5. ✓ Inference Pipeline
   - Built production-ready sentiment analyzer
   - Supports real-time predictions

6. ✓ Model Persistence
   - Model saved to: {os.path.basename(model_path)}
   - Configuration saved to: {os.path.basename(config_path)}
   - Verified model loading functionality

📊 KEY METRICS:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
• Total Parameters: {param_counts['total']:,}
• Trainable Parameters: {param_counts['trainable']:,} ({param_counts['trainable']/param_counts['total']*100:.2f}%)
• Parameter Reduction: {100 - param_counts['trainable']/param_counts['total']*100:.2f}%
• Best Validation Accuracy: {max(val_accuracies)*100:.2f}%
• Final Validation Accuracy: {val_accuracies[-1]*100:.2f}%
• Total Training Time: {total_training_time:.2f}s

💡 EFFICIENCY ANALYSIS:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
• Adapters achieved {max(val_accuracies)*100:.2f}% accuracy while training only 
  {param_counts['trainable']/1e6:.2f}M parameters (vs 110M for full fine-tuning)
• This represents a {100 - param_counts['trainable']/param_counts['total']*100:.1f}% 
  reduction in trainable parameters
• Memory savings: ~{(param_counts['frozen']/1e6):.0f}M parameters frozen
• Training time reduced significantly compared to full fine-tuning

📁 SAVED FILES:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
• Model checkpoint: {os.path.basename(model_path)}
• Configuration file: {os.path.basename(config_path)}
• Visualization: peft_adapter_complete_results.png

✅ ASSIGNMENT READY FOR SUBMISSION
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
""")

print("\n🎉 All cells executed successfully! Ready for university submission.")
print(f"📁 Model saved in '{model_dir}' folder")
print("🖼️ Visualization saved as 'peft_adapter_complete_results.png'")
```
